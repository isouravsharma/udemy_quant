```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
```

```python
stocks = ['ICICIBANK.NS']
ohlc_dict = {}
start = dt.datetime.today() - dt.timedelta(3650)
end = dt.datetime.today()
```

```python
for stock in stocks:
    ohlc_dict[stock] = yf.download(stock, start, end, interval='1d')
```

```python
# Calculate 50 DMA and 200 DMA
df = pd.DataFrame()
df['Close'] = ohlc_dict['ICICIBANK.NS']['Adj Close']
df['50DMA'] = df['Close'].rolling(50).mean()
df['200DMA'] = df['Close'].rolling(200).mean()
df['200_DIFF'] = df['Close'] - df['200DMA']
df['High_dff'] = df['200_DIFF'].max()
df['low_dff'] = df['200_DIFF'].min()
df['Below 200'] = np.where(df['Close'] <= df['200DMA'], 1, 0)
df
```
